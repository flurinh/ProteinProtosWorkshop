## The Scientific Story

We're continuing our exploration of the **Nature paper** (Zhao et al., 2023) on **PCO371** - the first orally available small-molecule agonist for Class B GPCRs.

**Key findings from the paper:**
- PCO371 binds at an *unexpected* site: the receptor-G protein interface
- The binding pocket is **conserved** across Class B GPCRs
- A single mutation (**P415A**) completely abolishes PCO371 activity
- Some mutations **enhance** activity (V412A, M414A)

**Our goal today:** Use ProtOS to analyze this data and design improved variants.

## Part 0: Understanding ProtOS

### The ProtOS Philosophy: Zero Configuration

Traditional bioinformatics workflows require:
- Setting up file paths
- Managing different file formats (PDB, mmCIF, FASTA...)
- Tracking which analyses have been run
- Converting between formats

**ProtOS handles all of this automatically.**

You only work with **human-readable names** like `"8jr9"` or `"pth1r_v412a"` - never with file paths.

````python
# Display the ProtOS architecture
from IPython.display import Image, display
display(Image(filename='protos/resources/overview2.png', width=900))
````

### The Processor Ecosystem

| Processor | Data Type | Key Operations |
|-----------|-----------|----------------|
| **StructureProcessor** | PDB/mmCIF | Download, load, filter, extract ligands |
| **SequenceProcessor** | FASTA | Read, align, mutate, conserve |
| **GRNProcessor** | Position tables | Map universal numbering, annotate |
| **LigandProcessor** | SDF/SMILES | Analyze interactions, properties |
| **EmbeddingProcessor** | Vectors | Generate ESM-2 embeddings |
| **PropertyProcessor** | Tables | Store experimental data |
| **ModelManager** | Model I/O | Prepare Boltz-2, RFdiffusion inputs |

**Data flows between processors** - a structure becomes sequences, sequences get GRN annotations, properties connect to embeddings, and everything feeds into models.

## Part 1: Easy Usage - Download & Explore Structures

Let's start simple: download the structures from the Nature paper.

````python
# Initialize ProtOS with workshop data directory
import sys
from pathlib import Path


# Add protos to path
sys.path.insert(0, 'protos/src')

# IMPORTANT: Configure data path BEFORE creating any processors
from protos.io.paths import ProtosPaths

# Set up workshop-specific data directory
workshop_data = Path('materials/session3/data').resolve()
workshop_data.mkdir(parents=True, exist_ok=True)

# Initialize ProtosPaths with our workshop directory
paths = ProtosPaths(data_root=str(workshop_data))
print(f"ProtOS data directory: {paths.data_root}")

# Now create processors - they will use our configured path
from protos.processing.structure import StructureProcessor

struct_proc = StructureProcessor(name="gpcr_workshop")
print(f"StructureProcessor initialized at: {struct_proc.paths.data_root}")
````

````python
# Define our Class B GPCR structures of interest
class_b_structures = {
    '8JR9': 'PCO371-PTH1R-Gs (from Nature paper)',
    '6NBF': 'LA-PTH-PTH1R-Gs (peptide reference)',
    '7LCI': 'GLP1R-Gs active state',
    '6X18': 'PTH2R-Gs',
}

print("Structures we'll analyze:")
for pdb_id, description in class_b_structures.items():
    print(f"  {pdb_id}: {description}")
````

````python
# Download structures - they're automatically registered!
from protos.io.ingest.download_structures import download_protein_structures

pdb_ids = list(class_b_structures.keys())

# Pass the processor so downloads go to our workshop directory
downloaded = download_protein_structures(pdb_ids, processor=struct_proc)

print(f"\nDownloaded {len(downloaded)} structures")
print("These are now registered in ProtOS and accessible by name.")
````

````python
# Load the PCO371-bound structure by name (not path!)
pco371_structure = struct_proc.load_entity('8jr9')

print(f"Loaded 8JR9 - PCO371-PTH1R-Gs complex")
print(f"Total atoms: {len(pco371_structure)}")
print(f"Chains: {pco371_structure['auth_chain_id'].unique().tolist()}")

# Find ligands
hetatms = pco371_structure[pco371_structure['record_type'] == 'HETATM']
ligands = hetatms['res_name3l'].unique()
print(f"Ligands: {[l for l in ligands if l not in ['HOH', 'CL', 'NA']]}")
````

````python
# Create a dataset - group structures for batch analysis
struct_proc.create_dataset(
    name="class_b_gpcr_active",
    entities=[pdb.lower() for pdb in pdb_ids],
    metadata={
        "description": "Active Class B GPCR structures with Gs",
        "source": "PDB",
        "paper": "Zhao et al., Nature 2023"
    }
)

print("Created dataset: 'class_b_gpcr_active'")
print(f"Contains: {pdb_ids}")
````

### Easy Usage Summary

With ~10 lines of code, we:
1. Initialized ProtOS
2. Downloaded 4 structures from PDB
3. Loaded and inspected a structure
4. Created a named dataset

**No file paths. No format conversions. No configuration.**

## Part 2: Medium Usage - Sequences & Alignment

Now let's extract sequences and find what makes PTH1R special for PCO371 binding.

````python
# Extract sequences from all structures
sequences = struct_proc.collect_chain_sequences([pdb.lower() for pdb in pdb_ids])

print("Extracted sequences:")
for pdb_id, chains in sequences.items():
    for chain_id, data in chains.items():
        seq_len = len(data['sequence'])
        print(f"  {pdb_id}_{chain_id}: {seq_len} residues")
````

````python
# Initialize SequenceProcessor and save receptor sequences
from protos.processing.sequence import SequenceProcessor

# Use the same paths configuration
seq_proc = SequenceProcessor(name="gpcr_workshop", paths=paths)

# Save receptor chains (typically the long ones, >300 residues)
receptor_sequences = {}
for pdb_id, chains in sequences.items():
    for chain_id, data in chains.items():
        if len(data['sequence']) > 300:  # Receptor chains are long
            name = f"{pdb_id}_receptor"
            receptor_sequences[name] = data['sequence']
            seq_proc.save_entity(name, data['sequence'])
            print(f"Saved: {name} ({len(data['sequence'])} aa)")
````

````python
# Create a sequence dataset
seq_proc.create_dataset(
    name="class_b_receptors",
    entities=list(receptor_sequences.keys()),
    metadata={"family": "Class B GPCR", "type": "receptor"}
)

print(f"Created sequence dataset with {len(receptor_sequences)} receptors")
````

````python
# Perform pairwise alignment to compare receptors
from protos.processing.sequence import align_blosum62, init_aligner, format_alignment

aligner = init_aligner()

# Compare PTH1R to GLP1R - why does one respond to PCO371 and not the other?
pth1r_seq = receptor_sequences.get('8jr9_receptor', list(receptor_sequences.values())[0])
glp1r_seq = receptor_sequences.get('7lci_receptor', list(receptor_sequences.values())[1])

alignment = align_blosum62(pth1r_seq[:100], glp1r_seq[:100], aligner)  # First 100 aa

print("PTH1R vs GLP1R alignment (first 100 residues):")
print(format_alignment(alignment))
````

## Part 3: Generic Residue Numbering (GRN)

### The Problem with Sequence Positions

Position **415** in PTH1R is NOT the same as position **415** in GLP1R!

Different receptors have:
- Different loop lengths
- Insertions and deletions
- Different numbering schemes

### The Solution: GRN

**Generic Residue Numbering** provides a universal coordinate system:
- `6.47b` means "position 47 in transmembrane helix 6, Class B numbering"
- This position is **P415** in PTH1R but **P370** in GLP1R
- Now we can compare equivalent positions across receptors!

````python
# Load the binding pocket data from Session 1
import pandas as pd

binding_pocket = pd.read_csv('materials/session1/solution/nature_figure_gpcr_b_by_hand.csv')
print("PCO371 binding pocket residues (from Session 1):")
display(binding_pocket)
````

### Key Observations from the Binding Pocket Table

**Conserved positions** (same amino acid across all receptors):
- `2x46`: R (Arginine) - hydrogen bonds with PCO371
- `2x50`: H (Histidine) - polar network
- `3x50`: E (Glutamate) - sulfonamide interaction
- `6x45`: L (Leucine) - hydrophobic pocket

**Variable positions** (differ between receptors):
- `3x47`: I in PTH1R, L in GLP1R
- `6x46`: M in PTH1R, I in GLP1R
- `6x47`: **P in PTH1R, L in PTH2R**  THIS IS THE KEY!
- `7x56`: I in PTH1R, L in GLP1R

````python
# The critical binding pocket residues and their GRN positions
pco371_pocket = {
    '2.46b': {'PTH1R': 'R219', 'role': 'H-bond with PCO371 head'},
    '2.50b': {'PTH1R': 'H223', 'role': 'Polar network'},
    '3.47b': {'PTH1R': 'I299', 'role': 'Hydrophobic contact'},
    '3.50b': {'PTH1R': 'E302', 'role': 'Sulfonamide interaction'},
    '6.45b': {'PTH1R': 'L413', 'role': 'Hydrophobic pocket'},
    '6.47b': {'PTH1R': 'P415', 'role': 'CRITICAL - selectivity switch'},
    '6.49b': {'PTH1R': 'F417', 'role': 'Pi-stacking'},
    '7.56b': {'PTH1R': 'I458', 'role': 'Hydrophobic contact'},
    '7.57b': {'PTH1R': 'Y459', 'role': 'Pi-stacking with phenyl'},
}

print("PCO371 binding pocket in PTH1R:")
print("="*60)
for grn, info in pco371_pocket.items():
    print(f"{grn:8}  {info['PTH1R']:6} : {info['role']}")
````

## Part 4: Connect Experimental Data

Now let's connect the structural analysis to the experimental data from Session 2.

````python
# Load the dose-response data from Session 2
selectivity_df = pd.read_csv('materials/session2/data/nature_figure_gpcr_b_selectivity.csv')

print("Dose-response data (PCO371):")
pco371_data = selectivity_df[selectivity_df['Ligand'] == 'PCO371']
print(pco371_data['Protein'].unique())
````

````python
# Display the experimental figure
from IPython.display import Image
display(Image(filename='materials/session2/data/nature_gpcr_b_abc_figure.jpg', width=700))
````

````python
# Create the mutant activity table (Extended Data Table 2 from paper)
mutant_data = pd.DataFrame({
    'Mutant': ['WT', 'R219A', 'H223A', 'L226A', 'I299A', 'E302A', 'L306A',
               'K408A', 'V412A', 'L413A', 'M414A', 'P415A', 'L416A', 'F417A',
               'I458A', 'Y459A', 'C462A', 'N463A', 'G464A', 'E465A'],
    'GRN': ['WT', '2.46b', '2.50b', '2.53b', '3.47b', '3.50b', '3.54b',
            '6.40b', '6.44b', '6.45b', '6.46b', '6.47b', '6.48b', '6.49b',
            '7.56b', '7.57b', '7.60b', '8.47b', '8.48b', '8.49b'],
    'pEC50': [6.74, 6.23, 6.31, 6.07, 6.09, 5.85, 6.78, 6.63, 7.49, 6.01,
              7.34, None, 7.01, 7.14, 6.51, 6.38, 6.32, 6.63, 7.07, 6.76],
    'PCO371_pct_WT': [100, 46.17, 40.93, 32.35, 60.52, 47.56, 78.32, 77.45,
                      97.10, 72.94, 88.78, 96.36, 75.75, 89.05, 80.93, 94.38,
                      86.41, 103.10, 100.36, 90.05],
    'Surface_Expression': [100, None, None, None, None, None, None, None,
                           None, None, None, 96.36, None, None, None, None,
                           None, None, None, None]
})

print("PTH1R Mutant Activity (PCO371):")
display(mutant_data)
````

````python
# Analyze the mutant data
print("\n=== MUTANT ANALYSIS ===")

# P415A is special - completely inactive!
p415a = mutant_data[mutant_data['Mutant'] == 'P415A']
print(f"\nP415A (6.47b): pEC50 = {p415a['pEC50'].values[0]}")
print("   COMPLETELY ABOLISHES PCO371 ACTIVITY")
print("   Surface expression normal (96.36%)")
print("   The proline at 6.47b is ESSENTIAL for PCO371 binding!")

# Find mutations that ENHANCE activity (pEC50 > WT)
wt_pec50 = 6.74
enhanced = mutant_data[(mutant_data['pEC50'].notna()) & (mutant_data['pEC50'] > wt_pec50)]

print(f"\nMutations that ENHANCE PCO371 potency (pEC50 > {wt_pec50}):")
for _, row in enhanced.iterrows():
    print(f"  {row['Mutant']:8} ({row['GRN']:6}): pEC50 = {row['pEC50']:.2f} (+{row['pEC50']-wt_pec50:.2f})")
````

### Key Insight from Mutant Data

**Mutations that DECREASE activity** (disrupted binding):
- R219A, H223A, E302A  Polar interactions broken
- I299A, L413A  Hydrophobic contacts disrupted
- **P415A  DEAD** (selectivity switch)

**Mutations that INCREASE activity** (improved binding!):
- **V412A** (pEC50: 7.49)  +0.75 units = ~5.6x more potent!
- **M414A** (pEC50: 7.34)  +0.60 units = ~4x more potent!
- L416A, F417A, G464A  Modest improvements

**Why do V412A and M414A enhance activity?**

These residues are adjacent to the binding pocket. Making them smaller (ValAla, MetAla) creates MORE SPACE for PCO371 to bind. This suggests PCO371 is slightly too bulky for the wild-type pocket!

````python
# Store mutant data in ProtOS PropertyProcessor
from protos.processing.property import PropertyProcessor

# Use the same paths configuration
prop_proc = PropertyProcessor(name="gpcr_workshop", paths=paths)

# Import as property dataset
prop_proc.import_properties(
    name="pth1r_mutant_activity",
    data=mutant_data,
    key_column="Mutant"
)

print("Stored mutant activity data in ProtOS")
````

## Part 5: Rational Design with Boltz-2

Now we'll use our insights to design improved variants and prepare them for structure prediction.

### Design Strategy

**Goal 1: Design an enhanced PCO371 responder**
- Combine V412A + M414A for synergistic improvement
- Predict: Will the double mutant be even better?

**Goal 2: Convert GLP1R to respond to PCO371**
- GLP1R-2M (two mutations) gains PCO371 response in the paper
- Target the non-conserved positions in the binding pocket

**Goal 3: Use Boltz-2 to validate designs**
- Predict structures of mutants
- Check if binding pocket geometry is preserved

````python
# Initialize ModelManager
from protos.models.model_manager import ModelManager

# Use the same paths configuration
manager = ModelManager(paths=paths)

# List available models
print("Available models for structure prediction:")
for model in manager.list_models():
    print(f"  - {model}")
````

````python
# Get PTH1R wild-type sequence
pth1r_wt = seq_proc.load_entity("8jr9_receptor")
print(f"PTH1R WT sequence: {len(pth1r_wt)} amino acids")
print(f"First 60 aa: {pth1r_wt[:60]}...")
````

````python
# Design mutant sequences
# Note: Position numbers are relative to the structure (8JR9)

def apply_mutation(sequence, position, new_aa):
    """Apply a single point mutation to a sequence."""
    seq_list = list(sequence)
    old_aa = seq_list[position - 1]  # Convert to 0-indexed
    seq_list[position - 1] = new_aa
    return ''.join(seq_list), old_aa

# Beneficial mutations from the paper (positions in 8JR9 structure)
# V412 and M414 in the paper numbering
# We need to find the correct positions in our extracted sequence

# For demonstration, let's create the mutants:
mutations = {
    'PTH1R_V412A': {'position': 412, 'new_aa': 'A', 'enhancement': '+0.75 pEC50'},
    'PTH1R_M414A': {'position': 414, 'new_aa': 'A', 'enhancement': '+0.60 pEC50'},
}

print("Designed mutants:")
for name, mut in mutations.items():
    print(f"  {name}: {mut['enhancement']}")
````

````python
# Create and save mutant sequences
# (In a real workflow, we'd apply the mutations to the actual sequence)

# For demonstration, let's show the Boltz-2 preparation workflow:
print("\n=== BOLTZ-2 SUBMISSION PREPARATION ===")
print("\nFor each designed variant, we would:")
print("1. Apply mutation(s) to the WT sequence")
print("2. Save the mutant sequence in ProtOS")
print("3. Prepare Boltz-2 input files")
print("4. Generate the submission command")

# Example Boltz-2 configuration
boltz_config = {
    "recycling": 5,           # Number of recycling iterations
    "num_samples": 3,         # Number of structure samples
    "crop_size": 512,         # Maximum sequence length
    "device": "cuda",         # Use GPU
}

print(f"\nBoltz-2 configuration:")
for key, value in boltz_config.items():
    print(f"  {key}: {value}")
````

````python
# Show what the Boltz-2 command would look like
print("\n=== BOLTZ-2 COMMANDS ===")
print("\nTo predict the wild-type structure:")
print("  boltz predict --fasta pth1r_wt.fasta --recycling 5 --samples 3")

print("\nTo predict enhanced mutants:")
print("  boltz predict --fasta pth1r_v412a.fasta --recycling 5 --samples 3")
print("  boltz predict --fasta pth1r_m414a.fasta --recycling 5 --samples 3")
print("  boltz predict --fasta pth1r_v412a_m414a.fasta --recycling 5 --samples 3")

print("\n(ProtOS ModelManager generates these commands automatically)")
````

### Inverse Design: Engineer GLP1R to Respond to PCO371

From the paper:
- WT GLP1R does NOT respond to PCO371
- GLP1R-2M (two mutations at binding pocket) GAINS response
- This opens the door to **oral diabetes drugs** using PCO371-like molecules!

The two key mutations convert the GLP1R binding pocket to match PTH1R.

````python
# Compare binding pocket differences between PTH1R and GLP1R
pocket_comparison = pd.DataFrame({
    'GRN': ['3x47', '6x43', '6x44', '6x46', '6x47', '6x48', '6x49', '7x56', '7x57'],
    'PTH1R': ['I', 'L', 'V', 'M', 'P', 'L', 'F', 'I', 'Y'],
    'GLP1R': ['L', 'L', 'T', 'I', 'P', 'L', 'L', 'L', 'Y'],
    'Conserved': ['No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes']
})

print("Binding pocket comparison (PTH1R vs GLP1R):")
display(pocket_comparison)

# Non-conserved positions are targets for engineering!
print("\nNon-conserved positions (engineering targets):")
non_conserved = pocket_comparison[pocket_comparison['Conserved'] == 'No']
for _, row in non_conserved.iterrows():
    print(f"  {row['GRN']}: {row['GLP1R']}  {row['PTH1R']}")
````

### Key Scientific Insight

**P415^6.47b is the "master switch" for PCO371 selectivity**
- Proline creates a kink in TM6 that forms the binding pocket
- PTH2R has Leucine at this position  no kink  no PCO371 binding

**V412A and M414A ENHANCE activity**
