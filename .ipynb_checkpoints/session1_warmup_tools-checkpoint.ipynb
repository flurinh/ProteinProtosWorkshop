## 4. Control Flow & Loops

```python
threshold = 0.13
for idx, reading in enumerate(absorbance_readings, start=1):
    if reading > threshold:
        print(f'Reading {idx} high: {reading:.2f}')
    else:
        print(f'Reading {idx} ok: {reading:.2f}')
```

```python
count = 0
while count < 3:
    print('Loop iteration', count)
    count += 1
```

## 5. Functions & Errors

```python
def normalize_reading(value, baseline=0.1):
    """Subtract baseline and return normalized value."""
    return value - baseline

normalize_reading(0.18)
```

```python
def describe_sample(name: str, operator: str, temp_c: float = 22.0) -> str:
    return f'{name} prepared by {operator} at {temp_c}°C'

describe_sample('Opsin', 'Dee')
```
